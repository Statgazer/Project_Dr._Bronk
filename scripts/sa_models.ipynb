BERT BASED CLASSIFICATION MODEL nlptown/bert-base-multilingual-uncased-sentiment

```python
# IMPORTING MODULES AND CREATING THE SENTIMENT CLASSIFICATION FUNCTION
#--------------*******************************------------------------

import torch
from transformers import AutoTokenizer, AutoModelForSequenceClassification
from torch.nn.functional import softmax
import matplotlib.pyplot as plt

# Load the tokenizer and model
model_name = "nlptown/bert-base-multilingual-uncased-sentiment"
tokenizer = AutoTokenizer.from_pretrained(model_name)
model = AutoModelForSequenceClassification.from_pretrained(model_name)

def sentiment_score_decode(review):
    # Tokenize the review with truncation
    tokens = tokenizer.encode(review, truncation=True, return_tensors='pt')
    
    # # Decode tokens to text to see the processed part of the review
    # decoded_review = tokenizer.decode(tokens[0])
    # print("Processed Text:", decoded_review)

    # Perform sentiment analysis
    result = model(tokens)

    # Extract logits
    logits = result.logits
    # print("Logits:", logits)

    # Apply softmax to convert logits to probabilities
    probabilities = softmax(logits, dim=1)
    # print("Probabilities:", probabilities)

    # Convert the result to a sentiment class using argmax
    sentiment_class = torch.argmax(probabilities, dim=1).item() + 1
    print("Sentiment Class:", sentiment_class)
    
    #plotting
    plt.figure(figsize=(3,3))  # Smaller figure size
    bar_width = 0.4
    plt.bar(range(probabilities.size(1)), probabilities.detach().numpy()[0])
    plt.xlabel('Sentiment Classes')
    plt.ylabel('Probability')
    plt.title('Sentiment Analysis Probabilities')
    plt.show()

    return sentiment_class, logits, probabilities
```

```python
# IMPORTING MODULES AND CREATING THE SENTIMENT CLASSIFICATION FUNCTION
#--------------*******************************------------------------

import torch
from transformers import AutoTokenizer, AutoModelForSequenceClassification
from torch.nn.functional import softmax
import matplotlib.pyplot as plt

# Load the tokenizer and model
model_name = "distilbert-base-uncased-finetuned-sst-2-english"
tokenizer = AutoTokenizer.from_pretrained(model_name)
model = AutoModelForSequenceClassification.from_pretrained(model_name)

def sentiment_score_distilbert(review):
    # Tokenize the review with truncation
    tokens = tokenizer.encode(review, truncation=True, return_tensors='pt')
    

    # Perform sentiment analysis
    result = model(tokens)

    # Extract logits
    logits = result.logits
    # print("Logits:", logits)

    # Apply softmax to convert logits to probabilities
    probabilities = softmax(logits, dim=1)
    # print("Probabilities:", probabilities)

    negative_score, positive_score = probabilities[0].tolist()

    print(f"Positive Score: {positive_score:.2f}, Negative Score: {negative_score:.2f}\n")
    
```
